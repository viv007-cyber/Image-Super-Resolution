```python
import os
import glob
from torch.utils.data import Dataset
from PIL import Image
```

```python
from torchvision import transforms
from torch.utils.data import DataLoader

# Define transforms for LR and HR separately
lr_transform = transforms.Compose([
    transforms.Resize((64, 64)),  # Low-resolution input
    transforms.ToTensor(),
])

hr_transform = transforms.Compose([
    transforms.Resize((256, 256)),  # High-resolution ground truth
    transforms.ToTensor(),
])

# Update MySRDataset to accept two transforms
class MySRDataset(Dataset):
    def __init__(self, lr_dir, hr_dir, lr_transform=None, hr_transform=None):
        self.lr_paths = sorted(glob.glob(os.path.join(lr_dir, '*.png')))
        self.hr_paths = sorted(glob.glob(os.path.join(hr_dir, '*.png')))
        self.lr_transform = lr_transform
        self.hr_transform = hr_transform

    def __len__(self):
        return len(self.lr_paths)

    def __getitem__(self, idx):
        lr_image = Image.open(self.lr_paths[idx]).convert("RGB")
        hr_image = Image.open(self.hr_paths[idx]).convert("RGB")

        if self.lr_transform:
            lr_image = self.lr_transform(lr_image)
        if self.hr_transform:
            hr_image = self.hr_transform(hr_image)

        return lr_image, hr_image

# Now create the dataset using both transforms
train_dataset = MySRDataset(
    '/kaggle/input/superres/image_super_resolution_dataset/scaling_4x/train/LR',
    '/kaggle/input/superres/image_super_resolution_dataset/scaling_4x/train/HR',
    lr_transform=lr_transform,
    hr_transform=hr_transform
)

# Create the DataLoader
train_loader = DataLoader(train_dataset, batch_size=1, shuffle=True, num_workers=2)
```

```python
import lpips
import torch.nn as nn
from tqdm import tqdm

# Ensure model and metrics are on the correct device
model = model.to(device)
criterion_lpips = lpips.LPIPS(net='vgg').to(device)
criterion_mse = nn.MSELoss()

def normalize_for_lpips(x):
    return x * 2 - 1  # Convert from [0, 1] to [-1, 1]

num_epochs=5
model.train()

# Logging for plotting
epoch_lpips_losses = []
epoch_mse_losses = []
epoch_psnrs = []
epoch_ssims = []
epochs = list(range(1, num_epochs + 1))

for epoch in range(num_epochs):
    # Epoch-wide accumulators
    epoch_lpips_sum = 0.0
    epoch_mse_sum = 0.0
    epoch_psnr_sum = 0.0
    epoch_ssim_sum = 0.0

    print(f"\nEpoch [{epoch+1}/{num_epochs}]")
    for i, (lr, hr) in enumerate(tqdm(train_loader, desc=f"Training Epoch {epoch+1}")):
        lr = lr.to(device)
        hr = hr.to(device)

        optimizer.zero_grad()
        sr = model(lr)

        sr_lpips = normalize_for_lpips(sr)
        hr_lpips = normalize_for_lpips(hr)

        loss_lpips = criterion_lpips(sr_lpips, hr_lpips).mean()
        loss_mse = criterion_mse(sr, hr)
        total_loss =  loss_mse

        total_loss.backward()
        optimizer.step()

        # Accumulate for epoch
        epoch_lpips_sum += loss_lpips.item()
        epoch_mse_sum += loss_mse.item()

        sr_eval = torch.clamp(sr, 0, 1)
        hr_eval = torch.clamp(hr, 0, 1)

        psnr_val = psnr_metric(sr_eval, hr_eval)
        ssim_val = ssim_metric(sr_eval, hr_eval)
        epoch_psnr_sum += psnr_val.item()
        epoch_ssim_sum += ssim_val.item()

        # Optional stepwise reporting (every 10 batches)
        if (i + 1) % 10 == 0:
            print(f'Step [{i+1}/{len(train_loader)}], '
                  f'LPIPS Loss: {loss_lpips.item():.4f}, '
                  f'MSE Loss: {loss_mse.item():.4f}, '
                  f'PSNR: {psnr_val.item():.4f}, SSIM: {ssim_val.item():.4f}')

    # Compute epoch averages
    num_batches = len(train_loader)
    epoch_lpips_losses.append(epoch_lpips_sum / num_batches)
    epoch_mse_losses.append(epoch_mse_sum / num_batches)
    epoch_psnrs.append(epoch_psnr_sum / num_batches)
    epoch_ssims.append(epoch_ssim_sum / num_batches)

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 4))

plt.subplot(1, 4, 1)
plt.plot(epochs, epoch_lpips_losses, marker='o')
plt.title('LPIPS Loss')

plt.subplot(1, 4, 2)
plt.plot(epochs, epoch_mse_losses, marker='o')
plt.title('MSE Loss')

plt.subplot(1, 4, 3)
plt.plot(epochs, epoch_psnrs, marker='o')
plt.title('PSNR')

plt.subplot(1, 4, 4)
plt.plot(epochs, epoch_ssims, marker='o')
plt.title('SSIM')

plt.tight_layout()
plt.show()
```
